```python
import pandas as pd
import glob
import os
from collections import defaultdict

# Path to derived_metrics folder
csv_folder = ''
csv_files = glob.glob(f'*.csv')

# Dictionary to store valid counts per project_id
project_metric_counts = defaultdict(int)

for file in csv_files:
    try:
        df = pd.read_csv(file)

        # Standardize columns (some files have more columns)
        if 'value' not in df.columns or 'project_id' not in df.columns:
            continue

        # Filter out nulls and zeroes
        df_filtered = df[['project_id', 'value']].dropna()
        df_filtered = df_filtered[df_filtered['value'] != 0]

        # Count valid rows per project_id
        valid_counts = df_filtered['project_id'].value_counts()

        # Add to global counter
        for project_id, count in valid_counts.items():
            project_metric_counts[project_id] += count

    except Exception as e:
        print(f"Error reading {file}: {e}")

# Convert to DataFrame and sort
result_df = pd.DataFrame(list(project_metric_counts.items()), columns=['project_id', 'valid_metric_count'])
top_5_projects = result_df.sort_values(by='valid_metric_count', ascending=False).head(5)

print("Top 5 project_ids with highest non-null/non-zero metric coverage:")
print(top_5_projects)
```

```python
import pandas as pd
import glob
import os
from collections import defaultdict

# Set working directory path
csv_folder = '/content'
csv_files = glob.glob(f'{csv_folder}/*.csv')

# Required metrics that each project_id must appear in
REQUIRED_METRICS = {
    'preprocessed_token_terminal_fees.csv',
    'preprocessed_token_terminal_transaction_count.csv'
}

# Normalize required metric filenames (strip and lowercase)
REQUIRED_METRICS = {f.strip().lower() for f in REQUIRED_METRICS}

# Initialize tracking structures
project_metric_counts = defaultdict(int)
project_to_metrics = defaultdict(set)
file_to_project_ids = {}

# Load and process each CSV
for file in csv_files:
    try:
        df = pd.read_csv(file)

        # Only process files with necessary columns
        if 'project_id' not in df.columns or 'value' not in df.columns:
            continue

        # Clean data: drop NaNs and filter value != 0
        df_filtered = df[['project_id', 'value']].dropna()
        df_filtered = df_filtered[df_filtered['value'] != 0]

        # Normalize filename for consistency
        filename = os.path.basename(file).strip().lower()

        # Track which project_ids are present in this file
        valid_project_ids = set(df_filtered['project_id'].unique())
        file_to_project_ids[filename] = valid_project_ids

        # Update counts and metrics per project_id
        counts = df_filtered['project_id'].value_counts()
        for project_id, count in counts.items():
            project_metric_counts[project_id] += count
            project_to_metrics[project_id].add(filename)

    except Exception as e:
        print(f"Error processing {file}: {e}")

# === Debug: Show available files and compare ===
print("✅ Available metric files:")
for fname in sorted(file_to_project_ids.keys()):
    print("-", fname)

print("\n🔍 Required metric files:")
for fname in sorted(REQUIRED_METRICS):
    print("-", fname)

# === Step 1: Check if all required metric files are available ===
missing_required = REQUIRED_METRICS - set(file_to_project_ids.keys())
if missing_required:
    raise ValueError(f"❌ Missing required files: {missing_required}")

# === Step 2: Filter only those project_ids that appear in all required files ===
eligible_projects = set.intersection(*(file_to_project_ids[metric] for metric in REQUIRED_METRICS))

# === Step 3: Create DataFrame and get top 5 projects ===
filtered_df = pd.DataFrame([
    (pid, project_metric_counts[pid]) for pid in eligible_projects
], columns=['project_id', 'valid_metric_count'])

top_5 = filtered_df.sort_values(by='valid_metric_count', ascending=False).head(5)

# === Step 4: Intersect metrics across top 5 project_ids ===
common_metrics = set.intersection(*(project_to_metrics[pid] for pid in top_5['project_id']))

# === Step 5: Output Results ===
print("\n🎯 Top 5 project_ids (present in both required metrics):")
print(top_5.to_string(index=False))

print("\n📊 Common metrics (CSV files) across these top 5 projects:")
for metric in sorted(common_metrics):
    print(f"- {metric}")
project_id_selected_list=list(top_5['project_id'])
print(project_id_selected_list)
```

```python
import os
import pandas as pd
from pathlib import Path

# === Path to your CSV directory ===
data_dir = "selected_projects"

# === Helper to extract metric name from filename ===
def extract_metric_name(filename):
    name = Path(filename).stem
    name = name.replace("preprocessed_token_terminal_", "")
    name = name.replace("token_", "")  # Optional: strip 'token_' if present
    return name

# === Main processing loop ===
for file in os.listdir(data_dir):
    if not file.endswith(".csv"):
        continue

    file_path = os.path.join(data_dir, file)
    metric_name = extract_metric_name(file)

    try:
        df = pd.read_csv(file_path)
    except pd.errors.ParserError:
        try:
            df = pd.read_csv(file_path, sep=';')
        except Exception as e:
            print(f"❌ Error reading {file}: {e}")
            continue

    # === Case 1: Rename 'value' to metric name ===
    if "value" in df.columns:
        df = df.rename(columns={"value": metric_name})
        df.to_csv(file_path, index=False)
        print(f"✅ Renamed 'value' → '{metric_name}' in {file}")

    # === Case 2: Already renamed ===
    elif metric_name in df.columns:
        print(f"ℹ️ Already renamed in {file} (has '{metric_name}')")

    # === Case 3: Unexpected structure ===
    else:
        print(f"⚠️ Skipping {file} — expected 'value' or '{metric_name}', found: {df.columns.tolist()}")
```

```python
import pandas as pd
import os
from pathlib import Path
import json
import yaml

# === CONFIG ===
TOKEN_LIST = project_id_selected_list  # project_ids
INPUT_DIR = '/content/selected_projects'
OUTPUT_ROOT = './project_root/data'
SNAPSHOT_CUTOFF = '2024-06'
SNAPSHOT_CUTOFF_MAX = '2025-06'

# === Derived file logic ===
def is_derived_file(filename):
    return not filename.startswith('preprocessed_')

# === Create output folders ===
Path(OUTPUT_ROOT).mkdir(parents=True, exist_ok=True)
for subdir in ['tokens', 'benchmarks', 'meta']:
    Path(f"{OUTPUT_ROOT}/{subdir}").mkdir(parents=True, exist_ok=True)

# === Save token_map.json ===
token_map = {token.lower(): token.upper() for token in TOKEN_LIST}
with open(f"{OUTPUT_ROOT}/meta/token_map.json", "w") as f:
    json.dump(token_map, f, indent=2)

# === Initialize data stores ===
metrics_mapping = {}
token_snapshots = {token: [] for token in TOKEN_LIST}
token_archives = {token: [] for token in TOKEN_LIST}
token_derived = {token: [] for token in TOKEN_LIST}
metrics_seen_per_token = {token: set() for token in TOKEN_LIST}
skipped_merges = []

# === Process CSV files ===
for file in os.listdir(INPUT_DIR):
    if not file.endswith('.csv'):
        continue

    file_path = os.path.join(INPUT_DIR, file)
    try:
        df = pd.read_csv(file_path)
    except pd.errors.ParserError:
        try:
            df = pd.read_csv(file_path, sep=';')
        except:
            df = pd.read_csv(file_path, sep='\t')

    print(f"📁 File: {file}")
    print(f"   Columns: {df.columns.tolist()}")
    print(f"   Shape: {df.shape}")
    if not df.empty:
        print(f"   Sample data: {df.iloc[0].to_dict()}")
    print()

    if 'year_month' not in df.columns or 'project_id' not in df.columns:
        print(f"⚠️ Skipping {file} - missing 'year_month' or 'project_id'")
        continue

    # Drop unnecessary columns
    df = df.drop(columns=[col for col in ['timestamp', 'project_name'] if col in df.columns])
    df['year_month'] = df['year_month'].astype(str)

    # Handle Token Terminal format
    if 'metric_id' in df.columns and 'value' in df.columns:
        metric_name = file.replace('preprocessed_token_terminal_', '').replace('.csv', '')
        df = df.rename(columns={'value': metric_name})
        df = df.drop(columns=['metric_id'])
        metric_cols = [metric_name]

    elif 'metric_id' in df.columns:
        metric_cols = [col for col in df.columns if col not in ['project_id', 'year_month', 'metric_id']]
        if len(metric_cols) == 1:
            metric_name = metric_cols[0]
            df = df.drop(columns=['metric_id'])
        else:
            print(f"⚠️ Skipping {file} - ambiguous columns: {metric_cols}")
            continue

    else:
        metric_cols = [col for col in df.columns if col not in ['project_id', 'year_month']]
        if len(metric_cols) == 1:
            metric_name = metric_cols[0]
        else:
            print(f"⚠️ Skipping {file} - ambiguous columns: {metric_cols}")
            continue

    # Aggregate if needed
    if df.duplicated(subset=['project_id', 'year_month']).any():
        print(f"⚠️ Aggregating multiple entries in {file} via mean.")
        df = df.groupby(['project_id', 'year_month']).mean(numeric_only=True).reset_index()

    # Determine metric type
    is_derived = is_derived_file(file)
    metric_type = "derived" if is_derived else "core"

    metrics_mapping[file] = {
        "metric_type": metric_type,
        "metric_name": metric_name,
        "columns": df.columns.tolist()
    }

    print(f"📊 Processing {file}: metric='{metric_name}', type='{metric_type}'")

    for token in TOKEN_LIST:
        df_token = df[df['project_id'].str.lower() == token.lower()].copy()
        if df_token.empty:
            continue

        if metric_name in metrics_seen_per_token[token]:
            print(f"⚠️ Skipping duplicate metric {metric_name} for {token}")
            continue
        metrics_seen_per_token[token].add(metric_name)

        df_token = df_token[['year_month', metric_name]].sort_values('year_month')
        df_snapshot = df_token[
                (df_token['year_month'] >= SNAPSHOT_CUTOFF) &
                (df_token['year_month'] < SNAPSHOT_CUTOFF_MAX)
            ]
        df_archive = df_token[df_token['year_month'] < SNAPSHOT_CUTOFF]

        if is_derived:
            df_derived = df_token[(df_token['year_month'] >= SNAPSHOT_CUTOFF) & (df_token['year_month'] < SNAPSHOT_CUTOFF_MAX)]
            if not df_derived.empty:
                token_derived[token].append(df_derived)
        else:
            if not df_snapshot.empty:
                token_snapshots[token].append(df_snapshot)
            if not df_archive.empty:
                token_archives[token].append(df_archive)

# === Merge helper ===
def merge_parts(parts, token_name):
    if not parts:
        return None

    df_merged = parts[0].copy()
    for i, df in enumerate(parts[1:], 1):
        if 'value' in df.columns:
            print(f"⚠️ Found leftover 'value' column in dataframe {i}, dropping it")
            df = df.drop(columns=['value'])

        overlapping_cols = set(df_merged.columns) & set(df.columns) - {'year_month'}
        if overlapping_cols:
            print(f"⚠️ Column conflict for {token_name}: {overlapping_cols}, skipping df #{i}")
            print("for skipped**********", list(df.columns))
            skipped_merges.append((token_name, i, overlapping_cols))
            continue

        df_merged = pd.merge(df_merged, df, on='year_month', how='outer')

    if 'value' in df_merged.columns:
        df_merged = df_merged.drop(columns=['value'])

    return df_merged.sort_values('year_month')

# === Cleanup existing outputs ===
for token in TOKEN_LIST:
    out_prefix = f"{OUTPUT_ROOT}/tokens/{token.upper()}_"
    for suffix in ["snapshot.csv", "archive.csv", "derived.csv"]:
        file_path = f"{out_prefix}{suffix}"
        if os.path.exists(file_path):
            os.remove(file_path)

# === Write output ===
for token in TOKEN_LIST:
    out_prefix = f"{OUTPUT_ROOT}/tokens/{token.upper()}_"

    merged_snapshot = merge_parts(token_snapshots[token], token)
    merged_archive = merge_parts(token_archives[token], token)
    merged_derived = merge_parts(token_derived[token], token)

    if merged_snapshot is not None:
        merged_snapshot.to_csv(out_prefix + "snapshot.csv", index=False)
        print(f"✅ Created {token.upper()}_snapshot.csv with columns: {list(merged_snapshot.columns)}")
        print(f"   Sample row: {merged_snapshot.iloc[0].to_dict() if not merged_snapshot.empty else 'No data'}")

    if merged_archive is not None:
        merged_archive.to_csv(out_prefix + "archive.csv", index=False)
        print(f"✅ Created {token.upper()}_archive.csv with columns: {list(merged_archive.columns)}")

    if merged_derived is not None:
        merged_derived.to_csv(out_prefix + "derived.csv", index=False)
        print(f"✅ Created {token.upper()}_derived.csv with columns: {list(merged_derived.columns)}")

    print()

# === Save metrics metadata ===
with open(f"{OUTPUT_ROOT}/meta/metrics.yaml", 'w') as f:
    yaml.dump(metrics_mapping, f, sort_keys=False)

# === Skipped Merges Summary ===
if skipped_merges:
    print("⚠️ Skipped Merges Summary:")
    for token, i, cols in skipped_merges:
        print(f"  {token} → df #{i} skipped due to overlapping columns: {cols}")

print("✅ Datalake structured with project_id and cleaned safely.")
```

```python
import pandas as pd
from pathlib import Path

# === CONFIG ===
DERIVED_DIR = Path("/content/selected_projects")
BENCHMARK_DIR = Path("project_root/data/benchmarks")
BENCHMARK_DIR.mkdir(parents=True, exist_ok=True)

# === METRIC FILE MAP ===
METRIC_FILE_MAP = {
    "dau_mau_ratio": "dau_mau_ratio.csv",
    "tokenholders_growth_rate": "tokenholder_growth.csv",
    "price_drawdown_max_90d": "price_drawdown_max_90d.csv",
    "price_volatility_30d": "price_volatility_30d.csv",
    "price_momentum_30d": "price_momentum_30d.csv",
    "sharpe_like_ratio": "sharpe_like_ratio.csv",
    "correlation_with_btc": "correlation_with_btc.csv",
    "fees_per_tx": "fees_per_tx.csv",
    "fdv_ps_ratio": "fdv_ps_ratio.csv"
}

# === GENERATE PEER COMPARISON CSV ===
def generate_peer_comparison_csv():
    metric_dfs = []
    for metric_id, filename in METRIC_FILE_MAP.items():
        file_path = DERIVED_DIR / filename
        df = pd.read_csv(file_path)

        # Standardize value column
        if 'value' not in df.columns:
            value_col = [col for col in df.columns if col not in {'project_id', 'timestamp', 'year_month', 'metric_id', 'project_name'}]
            df = df.rename(columns={value_col[0]: 'value'})

        df['year_month'] = pd.to_datetime(df['year_month'])

        df_metric = df[['project_id', 'year_month', 'value']].copy()
        df_metric['metric_id'] = metric_id
        metric_dfs.append(df_metric)

    # Combine all into long format
    peer_long_df = pd.concat(metric_dfs, ignore_index=True)

    # Add ranking and percentile per metric per month
    def rank_within_group(group):
        group = group.copy()
        group['rank'] = group['value'].rank(ascending=False, method='min')
        group['percentile'] = group['value'].rank(pct=True) * 100
        return group

    peer_ranked_df = peer_long_df.groupby(['year_month', 'metric_id'], group_keys=False).apply(rank_within_group)
    peer_ranked_df = peer_ranked_df[['year_month', 'project_id', 'metric_id', 'value', 'rank', 'percentile']]

    # Save output
    output_path = BENCHMARK_DIR / "token_peer_stats.csv"
    peer_ranked_df=peer_ranked_df[peer_ranked_df['project_id'].isin(project_id_selected_list)]
    peer_ranked_df=peer_ranked_df[(peer_ranked_df['year_month']>=SNAPSHOT_CUTOFF)&(peer_ranked_df['year_month']<SNAPSHOT_CUTOFF_MAX)]
    peer_ranked_df.to_csv(output_path, index=False)
    print(f"✅ Peer comparison file saved at {output_path}")

# === Run it ===
if __name__ == "__main__":
    generate_peer_comparison_csv()
```
